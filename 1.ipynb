```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle
```

```python
data=pd.read_csv('F:\data science\datasets\salary.csv')
```

```python
data
```

```python
data['experience'].fillna(0,inplace=True)
```

```python
data
```

```python
data['test_score'].fillna(data['test_score'].mean(), inplace=True)
```

```python
data
```

```python
X = data.iloc[:,:3]
```

```python
X['interview_score'].fillna(X['interview_score'].mean(), inplace=True)
```

```python
X
```

```python
def convert_to_int(word):
    word_dict = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'fifteen':15, 'six':6, 'seven':7, 'eight':8,
                'thirteen':13,'nine':9, 'ten':10, 'eleven':11, 'twelve':12, 'zero':0, 0: 0}
    return word_dict[word]

X['experience'] = X['experience'].apply(lambda x : convert_to_int(x))
```

```python
y=data.iloc[:,-1]
```

```python
xdata=X.to_numpy()
ydata=y.to_numpy()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

#Fitting model with trainig data
regressor.fit(xdata, ydata)

# Saving model to disk
pickle.dump(regressor, open('model.pkl','wb'))

# Loading model to compare the results
model = pickle.load(open('model.pkl','rb'))
print(model.predict([[2, 9, 6]]))
```
